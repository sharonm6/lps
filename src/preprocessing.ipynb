# Pre-Processing

```python
ts_match = r'(\d{4})-(\d{2})-(\d{2}) (\d{2}):(\d{2}):(\d{2}).\d'

contacts_df['time_to_reply'] = 0

for i, row in contacts_df.iterrows():
    contact_ts = row['ts_contact_at']

    reply_ts = row['ts_reply_at']

    if type(reply_ts) == float or contact_ts == reply_ts:
        continue

    c_match = re.match(ts_match, contact_ts)
    if c_match: 
        y,m,d,h,min,s = c_match.groups()

    r_match = re.match(ts_match, reply_ts)
    if r_match:
        yr,mr,dr,hr,minr,sr = r_match.groups()
        
    dt_contact = datetime.datetime(int(y), int(m), int(d), int(h), int(min), int(s))
    dt_reply = datetime.datetime(int(yr), int(mr), int(dr), int(hr), int(minr), int(sr))
    time_diff = dt_reply - dt_contact

    contacts_df.loc[i, 'time_to_reply'] = time_diff.total_seconds()
```

```python
# time from accepting to booking

contacts_df['time_to_book'] = 0

for i, row in contacts_df.iterrows():
    accept_ts = row['ts_accepted_at']

    booking_ts = row['ts_booking_at']

    if type(accept_ts) == float or type(booking_ts) == float or accept_ts == booking_ts:
        continue

    a_match = re.match(ts_match, accept_ts)
    if a_match: 
        y,m,d,h,min,s = a_match.groups()

    b_match = re.match(ts_match, booking_ts)
    if b_match:
        yb,mb,db,hb,minb,sb = b_match.groups()

    dt_accept = datetime.datetime(int(y), int(m), int(d), int(h), int(min), int(s))
    dt_booking = datetime.datetime(int(yb), int(mb), int(db), int(hb), int(minb), int(sb))
    time_diff = dt_booking - dt_accept

    print(time_diff.total_seconds())

    contacts_df.loc[i, 'time_to_book'] = time_diff.total_seconds()
```
